```python
"""
JEE DIAGNOSTIC ENGINE — UNIFIED PIPELINE
═══════════════════════════════════════════════════════════════════════════

INPUT FILES REQUIRED:
─────────────────────────────────────────────────────────────────────────
1. df_master (DataFrame, 4481 rows)
   Columns: question_id, topic, question_type, text_length
   Source: Your existing cleaned PYQ corpus

2. df_question_nodes (DataFrame, 5150 rows)
   Columns: question_id, node_name, cognitive_operation, cognitive_demand
   Source: Your existing analysis output

3. df_question_errors (DataFrame, 5522 rows)
   Columns: question_id, error_name, error_type, description
   Source: Your existing analysis output

4. df_question_concepts (DataFrame, 11784 rows)
   Columns: question_id, topic, concept_name
   Source: Your existing concept tagging output

5. prerequisite_graph (dict)
   Format: {target_topic: [source_topic, ...]}
   Source: layer_1_PN.txt loaded as dict

6. foundational_skills (list of dicts)
   Each: {skill_id, skill_name, description, failure_signature,
          dependent_topics, cognitive_demand_signature, diagnostic_question}
   Source: prerequisite_final.json + one-time labeling of demand signatures

7. student_responses (DataFrame, ~4500 rows per student)
   Columns: student_id, question_id, is_correct, time_taken_seconds,
            confidence_rating, test_timestamp
   Source: SYNTHETIC — to be generated

8. test_composition (DataFrame, ~600 rows per student)
   Columns: student_id, test_timestamp, question_id
   Source: SYNTHETIC — generated alongside student_responses
   Note: includes ALL questions assigned to a test (attempted + unattempted)

PIPELINE DEPENDENCY ORDER:
  Stage 0: Pre-filter eligible sets
  Stage 1: F1, F2, F3, F4, F7, F8
  Stage 2: F5, F6 (need F1 for context)
  Stage 3: F9, F10, F11
  Stage 4: F14, F16 (need F11 momentum)
  Stage 5: F17 (needs F5, F9, F11, F14, F16)
  Stage 6: F15, F18 (need F17)
"""

import pandas as pd
import numpy as np
from scipy import stats as scipy_stats
from dataclasses import dataclass, field
from typing import Optional


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 1 — ALL CONSTANTS
# ═══════════════════════════════════════════════════════════════════════════════

# Pre-filter
MIN_QUESTIONS = 10            # item needs ≥10 questions in corpus to be eligible

# Group 1
G1_MIN_ATTEMPTS       = 3
G1_WEAK_THRESHOLD     = 0.50
G1_CONCEPT_FAIL_TH    = 0.60
G1_ERROR_FREQ_TH      = 5

# Group 2
G2_MIN_ATTEMPTS_BASIC     = 3
G2_MIN_ATTEMPTS_BEHAV     = 5
G2_RUSHED_FRACTION        = 0.40
G2_MCQ_BASELINE           = 90
G2_NUM_BASELINE           = 120
G2_OVERCONF_GAP           = 0.20
G2_OVERCONF_ACC_CEIL      = 0.60
G2_OVERCONF_PROC_FRAC     = 0.60
G2_OVERCONF_MIN_CONF      = 2.0
G2_UNDERCONF_GAP          = 0.20
G2_UNDERCONF_ACC_FLOOR    = 0.60
G2_UNDERCONF_MCQ_WARN     = 0.80
G2_UNDERCONF_SLOW_MULT    = 1.30
G2_UNDERCONF_RECENCY_CONF = 2.5
G2_UNDERCONF_RECENCY_N    = 2

# Group 3
G3_F8_MIN_ATTEMPTS        = 3
G3_F8_FOCUS_CEIL          = 0.35
G3_F8_UNEXPECTED_FLOOR    = 0.65
G3_F8_STRETCH_DIFF        = 0.75
G3_F9_WEAK_TOPIC          = 0.40
G3_F9_WEAK_PREREQ         = 0.40
G3_F9_HIGH_DEMAND_FRAC    = 0.60
G3_F10_MIN_TOPIC_ATTEMPTS = 10
G3_F10_WEAK_TOPIC         = 0.40
G3_F10_OVERLAP_RATIO      = 0.60
G3_F10_MIN_WEAK           = 2
G3_F10_LOW_ACC_CUTOFF     = 0.35

# Group 4
G4_BASELINE_TESTS         = 10
G4_CURRENT_TESTS          = 5
G4_HOT_DELTA  = 0.10   # was 0.15
G4_COLD_DELTA = 0.10   # was 0.15

G4_COLD_BASELINE_FLOOR    = 0.60
G4_STAGNANT_CURRENT_MAX   = 0.50

# Group 5
G5_MCQ_BASELINE           = 90
G5_NUM_BASELINE           = 120
G5_PHASE1_ACC             = 0.65
G5_PHASE2_ACC             = 0.40
G5_SKIP_ACC               = 0.25
G5_TRAGIC_EV              = 2.0
G5_TRAGIC_DIFF_CEIL       = 0.60
G5_GOOD_SKIP_EV           = 0.5
G5_GOOD_SKIP_DIFF         = 0.75

# Group 6
G6_MAINT_ACC_FLOOR        = 0.65
G6_MAINT_STD_CEIL         = 0.15
G6_MAINT_MIN_TESTS        = 5
G6_NOT_SAFE_ACC           = 0.60
G6_SEVERITY_W             = 2.0
G6_PREREQ_W               = 1.5
G6_STAGNANT_W             = 1.0
G6_OVERCONF_W             = 1.0
G6_TIPPING_W              = 1.0

# Shared
HIGH_TIER_DEMANDS = {'Build Then Solve', 'Build and Work Backwards', 'Reverse Engineer'}
ERROR_TYPE_PRIORITY = {'Conceptual error': 0, 'Setup error': 1, 'Procedural error': 2}
DEMAND_LEVELS = ['Direct Application', 'Formula with Judgment', 'Build Then Solve',
                 'Build and Work Backwards', 'Reverse Engineer']


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 2 — UTILITY FUNCTIONS
# ═══════════════════════════════════════════════════════════════════════════════

def build_eligible_sets(df_question_nodes, df_question_errors, df_question_concepts):
    """Pre-filter: only items with ≥MIN_QUESTIONS questions are eligible."""
    nc = df_question_nodes.groupby('node_name')['question_id'].nunique()
    eligible_nodes = set(nc[nc >= MIN_QUESTIONS].index)
    node_lookup = (df_question_nodes[df_question_nodes['node_name'].isin(eligible_nodes)]
        .drop_duplicates(subset='node_name')
        .set_index('node_name')[['cognitive_operation', 'cognitive_demand']]
        .to_dict(orient='index'))

    cc = df_question_concepts.groupby('concept_name')['question_id'].nunique()
    eligible_concepts = set(cc[cc >= MIN_QUESTIONS].index)

    ec = df_question_errors.groupby('error_name')['question_id'].nunique()
    eligible_errors = set(ec[ec >= MIN_QUESTIONS].index)
    error_lookup = (df_question_errors[df_question_errors['error_name'].isin(eligible_errors)]
        .drop_duplicates(subset='error_name')
        .set_index('error_name')[['error_type', 'description']]
        .to_dict(orient='index'))

    return eligible_nodes, eligible_concepts, eligible_errors, node_lookup, error_lookup


def compute_difficulty_proxy(df_master, df_question_nodes):
    df = df_master.copy()
    df['length_pctile'] = df['text_length'].rank(pct=True)
    df['type_score'] = df['question_type'].map({'MCQ': 0.3, 'Numerical': 0.7})
    dm = {'Direct Application': 0.1, 'Formula with Judgment': 0.3,
          'Build Then Solve': 0.6, 'Build and Work Backwards': 0.8, 'Reverse Engineer': 1.0}
    nd = df_question_nodes[['question_id', 'cognitive_demand']].drop_duplicates()
    df = df.merge(nd, on='question_id', how='left')
    df['demand_score'] = df['cognitive_demand'].map(dm).fillna(0.5)
    df['difficulty_proxy'] = (df['length_pctile']*0.3 + df['type_score']*0.3 + df['demand_score']*0.4).round(3)
    return df[['question_id', 'difficulty_proxy']]


def compute_topic_accuracy(sr, df_master):
    m = sr.merge(df_master[['question_id', 'topic']], on='question_id', how='inner')
    return m.groupby('topic')['is_correct'].mean().to_dict()


def score_question(is_correct, question_type):
    if is_correct: return 4
    return -1 if question_type == 'MCQ' else 0


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 3 — GROUP 1: LOCATING WEAKNESS (F1, F2, F3)
# ═══════════════════════════════════════════════════════════════════════════════

def compute_f1_node_weakness(sr, df_qn, eligible_nodes, node_lookup):
    """F1: Node-level weakness map. Cross-test."""
    merged = sr.merge(df_qn[['question_id', 'node_name']], on='question_id', how='inner')
    merged = merged[merged['node_name'].isin(eligible_nodes)]
    stats = merged.groupby('node_name').agg(
        attempts=('is_correct', 'count'), correct=('is_correct', 'sum')).reset_index()
    stats['accuracy'] = (stats['correct'] / stats['attempts']).round(3)
    def status(r):
        if r['attempts'] < G1_MIN_ATTEMPTS: return 'insufficient_data'
        return 'weak' if r['accuracy'] < G1_WEAK_THRESHOLD else 'not_flagged'
    stats['status'] = stats.apply(status, axis=1)
    stats['cognitive_operation'] = stats['node_name'].map(
        lambda n: node_lookup.get(n, {}).get('cognitive_operation', ''))
    stats['cognitive_demand'] = stats['node_name'].map(
        lambda n: node_lookup.get(n, {}).get('cognitive_demand', ''))
    so = {'weak': 0, 'not_flagged': 1, 'insufficient_data': 2}
    stats['_s'] = stats['status'].map(so)
    return stats.sort_values(['_s', 'accuracy']).drop(columns='_s').reset_index(drop=True)


def compute_f2_concept_precision(sr, df_qc, eligible_concepts, weak_nodes, df_qn):
    """F2: Concept-level drill within weak nodes. Cross-test."""
    wq = df_qn[df_qn['node_name'].isin(weak_nodes)][['question_id', 'node_name']].drop_duplicates()
    wr = sr.merge(wq, on='question_id', how='inner')
    wc = wr.merge(df_qc[['question_id', 'concept_name', 'topic']], on='question_id', how='inner')
    wc = wc[wc['concept_name'].isin(eligible_concepts)]
    if wc.empty: return pd.DataFrame()
    stats = wc.groupby(['node_name', 'concept_name', 'topic']).agg(
        total=('is_correct', 'count'), correct=('is_correct', 'sum'),
        wrong=('is_correct', lambda x: (~x).sum())).reset_index()
    stats['failure_rate'] = (stats['wrong'] / stats['total']).round(3)
    stats['flagged'] = (stats['failure_rate'] > G1_CONCEPT_FAIL_TH) & (stats['total'] >= G1_MIN_ATTEMPTS)
    return stats.sort_values(['node_name', 'flagged', 'failure_rate'],
                              ascending=[True, False, False]).reset_index(drop=True)


def compute_f3_error_patterns(sr, df_qe, eligible_errors, error_lookup, df_master):
    """F3: Recurring error patterns. Cross-test, wrong answers only."""
    wrong = sr[sr['is_correct'] == False]
    if wrong.empty: return pd.DataFrame()
    we = wrong.merge(df_qe[['question_id', 'error_name', 'error_type', 'description']],
                     on='question_id', how='inner')
    we = we.merge(df_master[['question_id', 'topic']], on='question_id', how='left')
    we = we[we['error_name'].isin(eligible_errors)]
    if we.empty: return pd.DataFrame()

    freq = we.groupby(['error_name', 'error_type', 'description']).agg(
        frequency=('question_id', 'count'),
        unique_qs=('question_id', 'nunique')).reset_index()

    tb = we.groupby(['error_name', 'topic']).size().reset_index(name='count')
    td = tb.groupby('error_name').apply(
        lambda g: dict(zip(g['topic'], g['count'])), include_groups=False).to_dict()
    freq['topics_breakdown'] = freq['error_name'].map(td)
    freq['topic_count'] = freq['topics_breakdown'].apply(lambda d: len(d) if isinstance(d, dict) else 0)

    recurring = freq[freq['frequency'] >= G1_ERROR_FREQ_TH].copy()
    tp = {'Setup error': 0, 'Procedural error': 1, 'Conceptual error': 2}
    recurring['_s'] = recurring['error_type'].map(tp)
    return recurring.sort_values(['_s', 'frequency'], ascending=[True, False]).drop(
        columns='_s').reset_index(drop=True)


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 4 — GROUP 2: CLASSIFYING FAILURE (F4, F5, F6, F7)
# ═══════════════════════════════════════════════════════════════════════════════

def compute_f4_failure_modes(sr, df_qe, df_master):
    """F4: Label each wrong answer as Concept Gap / Setup Gap / Execution Slip / Rushed."""
    wrong = sr[sr['is_correct'] == False].copy()
    if wrong.empty: return pd.DataFrame(), pd.DataFrame()
    wrong = wrong.merge(df_master[['question_id', 'question_type']], on='question_id', how='left')
    wrong['baseline'] = wrong['question_type'].map(
        {'MCQ': G2_MCQ_BASELINE, 'Numerical': G2_NUM_BASELINE}).fillna(G2_MCQ_BASELINE)
    wrong['is_rushed'] = wrong['time_taken_seconds'] < (wrong['baseline'] * G2_RUSHED_FRACTION)

    qet = (df_qe.groupby('question_id')['error_type']
        .apply(lambda t: min(t, key=lambda x: ERROR_TYPE_PRIORITY.get(x, 99)))
        .reset_index().rename(columns={'error_type': 'dom_error'}))
    wrong = wrong.merge(qet, on='question_id', how='left')
    wrong = wrong.dropna(subset=['dom_error'])

    def classify(r):
        if r['is_rushed']: return 'Rushed'
        if r['dom_error'] == 'Conceptual error': return 'Concept Gap'
        if r['dom_error'] == 'Setup error': return 'Setup Gap'
        if r['dom_error'] == 'Procedural error':
            return 'Execution Slip' if r['confidence_rating'] >= 2 else 'Concept Gap'
        return 'Concept Gap'
    wrong['failure_mode'] = wrong.apply(classify, axis=1)

    detail = wrong[['question_id', 'time_taken_seconds', 'confidence_rating',
                     'question_type', 'dom_error', 'failure_mode']].reset_index(drop=True)
    summary = detail.groupby('failure_mode').size().reset_index(name='count')
    summary['pct'] = (summary['count'] / summary['count'].sum() * 100).round(1)
    mo = {'Concept Gap': 0, 'Setup Gap': 1, 'Execution Slip': 2, 'Rushed': 3}
    summary['_s'] = summary['failure_mode'].map(mo)
    summary = summary.sort_values('_s').drop(columns='_s').reset_index(drop=True)
    return detail, summary


def compute_f5_overconfidence(sr, df_qn, df_qe, node_lookup):
    """F5: Overconfidence detection. Error-aware, min confidence floor."""
    merged = sr.merge(df_qn[['question_id', 'node_name']], on='question_id', how='inner')
    stats = merged.groupby('node_name').agg(
        attempts=('is_correct', 'count'), correct=('is_correct', 'sum'),
        avg_conf=('confidence_rating', 'mean')).reset_index()
    stats['accuracy'] = (stats['correct'] / stats['attempts']).round(3)
    stats['conf_norm'] = (stats['avg_conf'] / 3.0).round(3)
    stats['gap'] = (stats['conf_norm'] - stats['accuracy']).round(3)
    stats['flagged'] = ((stats['attempts'] >= G2_MIN_ATTEMPTS_BEHAV) &
        (stats['gap'] > G2_OVERCONF_GAP) & (stats['accuracy'] < G2_OVERCONF_ACC_CEIL) &
        (stats['avg_conf'] >= G2_OVERCONF_MIN_CONF))

    wrong_err = merged[merged['is_correct'] == False].merge(
        df_qe[['question_id', 'error_type']], on='question_id', how='inner')

    def diagnose(nn):
        ne = wrong_err[wrong_err['node_name'] == nn]
        if ne.empty: return 'mixed_overconfidence'
        tc = ne['error_type'].value_counts(normalize=True)
        if tc.get('Procedural error', 0) >= G2_OVERCONF_PROC_FRAC: return 'execution_overconfidence'
        if (tc.get('Conceptual error', 0) + tc.get('Setup error', 0)) >= G2_OVERCONF_PROC_FRAC:
            return 'concept_overconfidence'
        return 'mixed_overconfidence'

    stats['diagnosis'] = stats.apply(lambda r: diagnose(r['node_name']) if r['flagged'] else '', axis=1)
    stats['cognitive_operation'] = stats['node_name'].map(
        lambda n: node_lookup.get(n, {}).get('cognitive_operation', ''))
    return stats.sort_values(['flagged', 'gap'], ascending=[False, False]).reset_index(drop=True)


def compute_f6_underconfidence(sr, df_qn, df_master, node_lookup):
    """F6: Underconfidence detection. Reliability + recency aware."""
    merged = sr.merge(df_qn[['question_id', 'node_name', 'cognitive_demand']],
                      on='question_id', how='inner')
    merged = merged.merge(df_master[['question_id', 'question_type']], on='question_id', how='left')

    stats = merged.groupby('node_name').agg(
        attempts=('is_correct', 'count'), correct=('is_correct', 'sum'),
        avg_conf=('confidence_rating', 'mean')).reset_index()
    stats['accuracy'] = (stats['correct'] / stats['attempts']).round(3)
    stats['conf_norm'] = (stats['avg_conf'] / 3.0).round(3)
    stats['gap'] = (stats['accuracy'] - stats['conf_norm']).round(3)
    stats['flagged'] = ((stats['attempts'] >= G2_MIN_ATTEMPTS_BEHAV) &
        (stats['gap'] > G2_UNDERCONF_GAP) & (stats['accuracy'] >= G2_UNDERCONF_ACC_FLOOR))

    student_avg_time = sr['time_taken_seconds'].mean()
    correct_only = merged[merged['is_correct'] == True]
    node_qtypes = (df_qn[['question_id', 'node_name']].merge(
        df_master[['question_id', 'question_type']], on='question_id')
        .groupby('node_name')['question_type'].apply(lambda x: (x == 'Numerical').sum()).to_dict())

    def assess(nn):
        nc = correct_only[correct_only['node_name'] == nn]
        if nc.empty: return 'unreliable_strength', -1, 0, 0, '', False
        num_in_corpus = node_qtypes.get(nn, 0)
        mcq_frac = -1 if num_in_corpus == 0 else (nc['question_type'] == 'MCQ').mean()
        avg_t = nc['time_taken_seconds'].mean()
        t_ratio = avg_t / student_avg_time if student_avg_time > 0 else 1
        demand = nc['cognitive_demand'].mode()
        demand = demand.iloc[0] if len(demand) > 0 else ''

        node_all = merged[merged['node_name'] == nn].sort_values('test_timestamp')
        last_n = node_all.tail(G2_UNDERCONF_RECENCY_N)
        if last_n['confidence_rating'].mean() >= G2_UNDERCONF_RECENCY_CONF:
            return 'recently_corrected', mcq_frac, avg_t, t_ratio, demand, True

        if mcq_frac == -1:
            if demand in HIGH_TIER_DEMANDS or t_ratio > G2_UNDERCONF_SLOW_MULT:
                return 'costly_strength', mcq_frac, avg_t, t_ratio, demand, False
            return 'no_numerical_available', mcq_frac, avg_t, t_ratio, demand, False
        if mcq_frac >= G2_UNDERCONF_MCQ_WARN:
            return 'unreliable_strength', mcq_frac, avg_t, t_ratio, demand, False
        if demand in HIGH_TIER_DEMANDS or t_ratio > G2_UNDERCONF_SLOW_MULT:
            return 'costly_strength', mcq_frac, avg_t, t_ratio, demand, False
        return 'confident_strength', mcq_frac, avg_t, t_ratio, demand, False

    results = stats.apply(
        lambda r: assess(r['node_name']) if r['flagged'] else ('', 0, 0, 0, '', False), axis=1)
    stats['diagnosis'] = [r[0] for r in results]
    stats['mcq_frac'] = [round(r[1], 2) for r in results]
    stats['avg_correct_time'] = [round(r[2], 1) for r in results]
    stats['time_ratio'] = [round(r[3], 2) for r in results]
    stats['cognitive_demand'] = [r[4] for r in results]
    stats['recency_suppressed'] = [r[5] for r in results]
    stats['cognitive_operation'] = stats['node_name'].map(
        lambda n: node_lookup.get(n, {}).get('cognitive_operation', ''))
    return stats.sort_values(['flagged', 'gap'], ascending=[False, False]).reset_index(drop=True)


def compute_f7_demand_profile(sr, df_qn):
    """F7: Accuracy by cognitive demand level."""
    merged = sr.merge(df_qn[['question_id', 'cognitive_demand']].drop_duplicates(),
                      on='question_id', how='inner')
    ds = merged.groupby('cognitive_demand').agg(
        attempts=('is_correct', 'count'), correct=('is_correct', 'sum')).reset_index()
    ds['accuracy'] = (ds['correct'] / ds['attempts']).round(3)
    do = {l: i for i, l in enumerate(DEMAND_LEVELS)}
    ds['_o'] = ds['cognitive_demand'].map(do).fillna(99)
    ds = ds.sort_values('_o').drop(columns='_o')
    W = 20
    ds['bar'] = ds['accuracy'].apply(lambda a: '█' * int(a * W) + '░' * (W - int(a * W)))
    drop_off = None
    prev = None
    for _, r in ds.iterrows():
        if prev is not None and prev - r['accuracy'] >= 0.15:
            drop_off = r['cognitive_demand']; break
        prev = r['accuracy']
    return ds.reset_index(drop=True), drop_off


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 5 — GROUP 3: TRACING ROOT CAUSE (F8, F9, F10)
# ═══════════════════════════════════════════════════════════════════════════════

def compute_f8_classification(sr, df_qn, df_master, node_lookup, latest_date=None):
    """F8: Classify latest test wrong answers. 5 categories."""
    if latest_date is None: latest_date = sr['test_timestamp'].max()
    latest = sr[sr['test_timestamp'] == latest_date]
    historical = sr[sr['test_timestamp'] < latest_date]
    wrong = latest[latest['is_correct'] == False].copy()
    if wrong.empty: return pd.DataFrame()

    wrong = wrong.merge(df_qn[['question_id', 'node_name']], on='question_id', how='inner')
    diff = compute_difficulty_proxy(df_master, df_qn)
    wrong = wrong.merge(diff, on='question_id', how='left')

    if not historical.empty:
        hn = historical.merge(df_qn[['question_id', 'node_name']], on='question_id', how='inner')
        hs = hn.groupby('node_name').agg(
            hist_att=('is_correct', 'count'), hist_acc=('is_correct', 'mean')).reset_index()
        wrong = wrong.merge(hs, on='node_name', how='left')
        wrong['hist_att'] = wrong['hist_att'].fillna(0).astype(int)
    else:
        wrong['hist_att'] = 0; wrong['hist_acc'] = None

    def classify(r):
        if r['hist_att'] < G3_F8_MIN_ATTEMPTS: return 'no_baseline'
        if r['hist_acc'] <= G3_F8_FOCUS_CEIL: return 'focus_area'
        if r['hist_acc'] >= G3_F8_UNEXPECTED_FLOOR:
            return 'stretch_question' if r.get('difficulty_proxy', 0.5) >= G3_F8_STRETCH_DIFF else 'unexpected_slip'
        return 'developing'
    wrong['classification'] = wrong.apply(classify, axis=1)
    wrong['cognitive_operation'] = wrong['node_name'].map(
        lambda n: node_lookup.get(n, {}).get('cognitive_operation', ''))
    co = {'unexpected_slip': 0, 'stretch_question': 1, 'developing': 2, 'focus_area': 3, 'no_baseline': 4}
    wrong['_s'] = wrong['classification'].map(co)
    cols = ['question_id', 'node_name', 'hist_acc', 'hist_att', 'difficulty_proxy',
            'classification', 'cognitive_operation']
    return wrong.sort_values('_s').drop(columns='_s')[[c for c in cols if c in wrong.columns]].reset_index(drop=True)


def compute_f9_prerequisites(sr, df_master, df_qn, prereq_graph):
    """F9: Topic prerequisite trace with cognitive demand cross-check."""
    topic_acc = compute_topic_accuracy(sr, df_master)
    weak = [t for t, a in topic_acc.items() if a < G3_F9_WEAK_TOPIC]
    if not weak: return pd.DataFrame(), {}

    resp_topic = sr.merge(df_master[['question_id', 'topic']], on='question_id', how='inner')
    wrong_resp = resp_topic[resp_topic['is_correct'] == False]
    wrong_demand = wrong_resp.merge(
        df_qn[['question_id', 'cognitive_demand']].drop_duplicates(), on='question_id', how='inner')

    topic_hd = {}
    for t in weak:
        tw = wrong_demand[wrong_demand['topic'] == t]
        topic_hd[t] = tw['cognitive_demand'].isin(HIGH_TIER_DEMANDS).mean() if not tw.empty else 0

    rows = []; readiness = {}
    for wt in weak:
        prereqs = prereq_graph.get(wt, [])
        if not prereqs: continue
        hf = round(topic_hd.get(wt, 0), 2)
        wp = 1.0
        for pt in prereqs:
            pa = topic_acc.get(pt)
            if pa is None: status = 'no_data'
            elif pa < G3_F9_WEAK_PREREQ: status = 'also_weak'
            else: status = 'fine'
            if pa is not None and pa < wp: wp = pa
            tc = 'not_applicable'
            if status == 'also_weak':
                tc = 'confirmed' if hf >= G3_F9_HIGH_DEMAND_FRAC else ('suspected' if hf >= 0.40 else 'weak_signal')
            rows.append({'weak_topic': wt, 'weak_acc': round(topic_acc.get(wt, 0), 3),
                         'prereq_topic': pt, 'prereq_acc': round(pa, 3) if pa else None,
                         'prereq_status': status, 'high_demand_frac': hf, 'trace_confidence': tc})
        readiness[wt] = round(wp, 3)

    result = pd.DataFrame(rows)
    if not result.empty: result['readiness'] = result['weak_topic'].map(readiness)
    return result, readiness


def compute_f10_foundational(sr, df_master, df_qn, foundational_skills):
    """F10: Foundational skill trace — hypothesis mode."""
    resp_topic = sr.merge(df_master[['question_id', 'topic']], on='question_id', how='inner')
    ts = resp_topic.groupby('topic').agg(att=('is_correct', 'count'), acc=('is_correct', 'mean')).reset_index()
    attempted = set(ts[ts['att'] >= G3_F10_MIN_TOPIC_ATTEMPTS]['topic'])
    topic_acc = dict(zip(ts['topic'], ts['acc']))
    weak = set(ts[(ts['topic'].isin(attempted)) & (ts['acc'] <= G3_F10_WEAK_TOPIC)]['topic'])
    global_acc = resp_topic['is_correct'].mean()
    display = 'top_1' if global_acc <= G3_F10_LOW_ACC_CUTOFF else 'show_all'


    if not weak or not foundational_skills: return pd.DataFrame(), global_acc, display

    wrong_demand = resp_topic[resp_topic['is_correct'] == False].merge(
        df_qn[['question_id', 'cognitive_demand']].drop_duplicates(), on='question_id', how='inner')

    rows = []
    for skill in foundational_skills:
        dep = set(skill.get('dependent_topics', []))
        sig = set(skill.get('cognitive_demand_signature', []))
        att_dep = dep & attempted
        if not att_dep: continue
        overlap = weak & att_dep
        oc = len(overlap)
        ratio = oc / len(att_dep) if att_dep else 0

        if oc >= G3_F10_MIN_WEAK and ratio >= G3_F10_OVERLAP_RATIO:
            dm = None; conf = 'unverified'
            if sig and not wrong_demand.empty:
                mt = sum(1 for t in overlap
                         if not wrong_demand[wrong_demand['topic'] == t].empty and
                         wrong_demand[wrong_demand['topic'] == t]['cognitive_demand'].isin(sig).mean() >= 0.50)
                dm = mt / oc if oc > 0 else 0
                conf = 'strong_hypothesis' if dm >= 0.60 else ('moderate_hypothesis' if dm >= 0.30 else 'weak_hypothesis')

            pri = round(oc * np.mean([1 - topic_acc.get(t, 0.5) for t in overlap]) * (1 + (dm or 0)), 3)
            rows.append({
                'skill_id': skill['skill_id'], 'skill_name': skill['skill_name'],
                'description': skill.get('description', ''),
                'failure_signature': skill.get('failure_signature', ''),
                'weak_overlap': sorted(list(overlap)), 'overlap_count': oc,
                'attempted_dep': len(att_dep), 'overlap_ratio': round(ratio, 2),
                'demand_match': round(dm, 2) if dm is not None else None,
                'confidence': conf, 'priority': pri,
                'diagnostic_question': skill.get('diagnostic_question', ''),
            })

    result = pd.DataFrame(rows)
    if not result.empty: result = result.sort_values('priority', ascending=False)
    return result.reset_index(drop=True), global_acc, display


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 6 — GROUP 4: PERFORMANCE MOMENTUM (F11)
# ═══════════════════════════════════════════════════════════════════════════════

def compute_f11_momentum(sr, df_master):
    """F11: Performance Momentum — baseline (1-10) vs current form (11-15)."""
    merged = sr.merge(df_master[['question_id', 'topic', 'question_type']], on='question_id', how='inner')
    merged['marks'] = merged.apply(lambda r: score_question(r['is_correct'], r['question_type']), axis=1)

    # Overall trend
    test_scores = merged.groupby('test_timestamp').agg(
        net=('marks', 'sum'), qs=('question_id', 'count'),
        acc=('is_correct', 'mean')).reset_index().sort_values('test_timestamp')
    scores = test_scores['net'].tolist()
    x = np.arange(len(scores))
    slope = scipy_stats.linregress(x, scores).slope if len(scores) >= 3 else 0
    labels = {1.0: 'Steady Climb', 0.3: 'Slight Upward', -0.3: 'Flat', -1.0: 'Slight Decline'}
    tl = 'Declining'
    for thresh, label in sorted(labels.items(), reverse=True):
        if slope > thresh: tl = label; break
    cur_avg = test_scores.tail(G4_CURRENT_TESTS)['net'].mean()

    trend = {'scores': scores, 'slope': round(slope, 2), 'label': tl,
             'current_avg': round(cur_avg, 1),
             'max_possible': int(test_scores['qs'].iloc[0] * 4) if not test_scores.empty else 0}

    # Split baseline / current
    all_dates = sorted(test_scores['test_timestamp'].unique())
    bl_dates = set(all_dates[:G4_BASELINE_TESTS])
    cu_dates = set(all_dates[G4_BASELINE_TESTS:])
    topic_test = merged.groupby(['test_timestamp', 'topic']).agg(
        net=('marks', 'sum'), qs=('question_id', 'count'),
        correct=('is_correct', 'sum'), acc=('is_correct', 'mean')).reset_index()

    def phase_stats(dates):
        d = topic_test[topic_test['test_timestamp'].isin(dates)]
        return d.groupby('topic').agg(
            att=('qs', 'sum'), correct=('correct', 'sum'), marks=('net', 'sum'),
            tests=('test_timestamp', 'nunique')).reset_index()

    bl = phase_stats(bl_dates)
    bl['bl_acc'] = (bl['correct'] / bl['att']).round(3)
    bl['bl_mpt'] = (bl['marks'] / bl['tests']).round(1)
    cu = phase_stats(cu_dates)
    cu.columns = ['topic', 'cu_att', 'cu_correct', 'cu_marks', 'cu_tests']
    cu['cu_acc'] = (cu['cu_correct'] / cu['cu_att']).round(3)
    cu['cu_mpt'] = (cu['cu_marks'] / cu['cu_tests']).round(1)

    drivers = bl[['topic', 'att', 'bl_acc', 'bl_mpt']].merge(
        cu[['topic', 'cu_att', 'cu_acc', 'cu_mpt']], on='topic', how='outer')
    for c in ['att', 'cu_att']: drivers[c] = drivers[c].fillna(0).astype(int)
    drivers['delta'] = (drivers['cu_acc'] - drivers['bl_acc']).round(3)
    drivers['marks_delta'] = (drivers['cu_mpt'] - drivers['bl_mpt']).round(1)

    safety_bl = 4; safety_cu = 4
    drivers['eligible'] = (drivers['att'] >= safety_bl) & (drivers['cu_att'] >= safety_cu)

    def classify(r):
        if not r['eligible']: return 'insufficient_data'
        d, bl_a, cu_a = r['delta'], r['bl_acc'], r['cu_acc']
        if d >= G4_HOT_DELTA: return 'hot_streak'
        if d <= -G4_COLD_DELTA and bl_a >= G4_COLD_BASELINE_FLOOR: return 'cold_streak'
        if abs(d) < G4_HOT_DELTA and cu_a < G4_STAGNANT_CURRENT_MAX: return 'stagnant'
        return 'stable'

    drivers['momentum'] = drivers.apply(classify, axis=1)
    mo = {'hot_streak': 0, 'cold_streak': 1, 'stagnant': 2, 'stable': 3, 'insufficient_data': 4}
    drivers['_s'] = drivers['momentum'].map(mo)
    drivers = drivers.sort_values(['_s', 'delta'], ascending=[True, False]).drop(columns='_s')

    # Extract momentum_states dict for downstream features
    momentum_states = dict(zip(drivers['topic'], drivers['momentum']))

    return {'trend': trend, 'drivers': drivers.reset_index(drop=True), 'momentum_states': momentum_states}


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 7 — GROUP 5: EXAM STRATEGY (F14, F15)
# ═══════════════════════════════════════════════════════════════════════════════

def compute_f14_exam_plan(sr, df_master, df_qn):
    """F14: Phased exam plan + cognitive demand background."""
    merged = sr.merge(df_master[['question_id', 'topic', 'question_type']], on='question_id', how='inner')
    ts = merged.groupby('topic').agg(
        att=('is_correct', 'count'), acc=('is_correct', 'mean'),
        avg_time=('time_taken_seconds', 'mean')).reset_index()
    ts['acc'] = ts['acc'].round(3); ts['avg_time'] = ts['avg_time'].round(0).astype(int)
    n_tests = sr['test_timestamp'].nunique()
    ts['qpt'] = (ts['att'] / n_tests).round(1)

    def phase(r):
        if r['acc'] < G5_SKIP_ACC: return 'skip'
        if r['acc'] < G5_PHASE2_ACC: return 'phase_3'
        if r['acc'] >= G5_PHASE1_ACC and r['avg_time'] <= G5_MCQ_BASELINE * 1.1: return 'phase_1'
        return 'phase_2'
    ts['phase'] = ts.apply(phase, axis=1)
    ts['exp_marks'] = (ts['qpt'] * ts['acc'] * 4 - ts['qpt'] * (1-ts['acc']) * 0.7).round(1)

    pm = ts.groupby('phase')['exp_marks'].sum().to_dict()
    po = {'phase_1': 0, 'phase_2': 1, 'phase_3': 2, 'skip': 3}
    ts['_s'] = ts['phase'].map(po)
    ts = ts.sort_values(['_s', 'acc'], ascending=[True, False]).drop(columns='_s')

    # Cognitive demand analysis
    md = sr.merge(df_qn[['question_id', 'cognitive_demand']].drop_duplicates(), on='question_id', how='inner')
    da = md.groupby('cognitive_demand').agg(
        att=('is_correct', 'count'), acc=('is_correct', 'mean'),
        avg_t=('time_taken_seconds', 'mean')).reset_index()
    da['acc'] = da['acc'].round(3); da['avg_t'] = da['avg_t'].round(0).astype(int)

    # One-line insight
    ws = md[(md['is_correct'] == False) & (md['time_taken_seconds'] > G5_MCQ_BASELINE)]
    insight = "No clear cognitive demand pattern in time-drain questions."
    if not ws.empty:
        dc = ws['cognitive_demand'].value_counts()
        top_d = dc.index[0]; top_f = dc.iloc[0] / len(ws)
        if 'Build' in top_d or 'Reverse' in top_d:
            insight = (f"{top_f:.0%} of your time-drain questions are \"{top_d}\". "
                       "When you feel 'how do I even set this up?' — start the 3-min clock.")
        else:
            insight = (f"{top_f:.0%} of your time-drain questions are \"{top_d}\". "
                       "Slow down 30 seconds to identify what to plug in before computing.")

    proj = {p: round(pm.get(p, 0)) for p in ['phase_1', 'phase_2', 'phase_3']}
    proj['total'] = sum(proj.values())

    return {'phases': ts.reset_index(drop=True), 'demand': da.reset_index(drop=True),
            'insight': insight, 'projected': proj}


def compute_f15_time_roi(sr, df_master, df_qn, test_comp, latest_date=None):
    """F15: Time ROI for latest test — quadrants + unattempted analysis."""
    if latest_date is None: latest_date = sr['test_timestamp'].max()
    latest = sr[sr['test_timestamp'] == latest_date].copy()
    historical = sr[sr['test_timestamp'] < latest_date]
    if latest.empty: return {}

    latest = latest.merge(df_master[['question_id', 'topic', 'question_type']], on='question_id', how='inner')
    latest['marks'] = latest.apply(lambda r: score_question(r['is_correct'], r['question_type']), axis=1)
    latest['baseline'] = latest['question_type'].map({'MCQ': G5_MCQ_BASELINE, 'Numerical': G5_NUM_BASELINE}).fillna(G5_MCQ_BASELINE)
    latest['is_fast'] = latest['time_taken_seconds'] <= latest['baseline']

    def quad(r):
        if r['is_correct'] and r['is_fast']: return 'fast_correct'
        if r['is_correct'] and not r['is_fast']: return 'slow_correct'
        if not r['is_correct'] and r['is_fast']: return 'fast_wrong'
        return 'slow_wrong'
    latest['quadrant'] = latest.apply(quad, axis=1)

    qs = latest.groupby('quadrant').agg(
        questions=('question_id', 'count'),
        time_min=('time_taken_seconds', lambda x: round(x.sum()/60, 1)),
        marks=('marks', 'sum')).reset_index()
    qs['roi'] = (qs['marks'] / qs['time_min']).round(1)
    qo = {'fast_correct': 0, 'slow_correct': 1, 'fast_wrong': 2, 'slow_wrong': 3}
    qs['_s'] = qs['quadrant'].map(qo)
    qs = qs.sort_values('_s').drop(columns='_s')

    sw = latest[latest['quadrant'] == 'slow_wrong']
    sw_topics = pd.DataFrame()
    if not sw.empty:
        sw_topics = sw.groupby('topic').agg(
            time_min=('time_taken_seconds', lambda x: round(x.sum()/60, 1)),
            marks=('marks', 'sum'), count=('question_id', 'count')).reset_index().sort_values('time_min', ascending=False)

    # Unattempted
    test_qids = set(test_comp[test_comp['test_timestamp'] == latest_date]['question_id'])
    attempted_qids = set(latest['question_id'])
    unatt_qids = test_qids - attempted_qids
    unatt_analysis = []

    if unatt_qids and not historical.empty:
        ht = historical.merge(df_master[['question_id', 'topic']], on='question_id', how='inner')
        hta = ht.groupby('topic')['is_correct'].mean().to_dict()
        diff_d = dict(zip(*[compute_difficulty_proxy(df_master, df_qn)[c] for c in ['question_id', 'difficulty_proxy']]))

        for qid in unatt_qids:
            qi = df_master[df_master['question_id'] == qid]
            if qi.empty: continue
            t, qt = qi['topic'].values[0], qi['question_type'].values[0]
            d = diff_d.get(qid, 0.5); ta = hta.get(t, 0.3)
            ev = ta * 4 + (1 - ta) * (-1) if qt == 'MCQ' else ta * 4
            if ev >= G5_TRAGIC_EV and d < G5_TRAGIC_DIFF_CEIL: cls = 'tragic_miss'
            elif ev < G5_GOOD_SKIP_EV or d >= G5_GOOD_SKIP_DIFF: cls = 'good_skip'
            else: cls = 'borderline'
            unatt_analysis.append({'question_id': qid, 'topic': t, 'question_type': qt,
                                   'difficulty': round(d, 2), 'topic_acc': round(ta, 2),
                                   'ev': round(ev, 1), 'classification': cls})

    unatt_df = pd.DataFrame(unatt_analysis)
    score = latest['marks'].sum()
    leaked = latest[latest['is_correct'] == False]['marks'].sum()
    tragic_ev = unatt_df[unatt_df['classification'] == 'tragic_miss']['ev'].sum() if not unatt_df.empty else 0
    wrong_time = latest[latest['is_correct'] == False]['time_taken_seconds'].sum() / 60

    bl = {'score': int(score), 'leaked': int(leaked), 'without_leaks': int(score + abs(leaked)),
          'tragic_ev': round(tragic_ev, 0), 'with_tragic': int(score + abs(leaked) + tragic_ev),
          'total_time': round(latest['time_taken_seconds'].sum()/60, 1),
          'wrong_time': round(wrong_time, 1),
          'attempted': len(latest), 'unattempted': len(unatt_df)}

    return {'quadrants': qs.reset_index(drop=True), 'sw_topics': sw_topics.reset_index(drop=True),
            'unattempted': unatt_df, 'bottom_line': bl}


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 8 — GROUP 6: PRIORITIZING ACTION (F16, F17, F18)
# ═══════════════════════════════════════════════════════════════════════════════

def compute_f16_maintenance(sr, df_master, momentum_states):
    """F16: Maintenance mode — stable strong topics to deprioritize."""
    merged = sr.merge(df_master[['question_id', 'topic']], on='question_id', how='inner')
    overall = merged.groupby('topic')['is_correct'].mean().to_dict()
    all_dates = sorted(merged['test_timestamp'].unique())
    if len(all_dates) < G6_MAINT_MIN_TESTS: return pd.DataFrame()

    last_dates = all_dates[-G6_MAINT_MIN_TESTS:]
    recent = merged[merged['test_timestamp'].isin(last_dates)]
    pta = recent.groupby(['topic', 'test_timestamp'])['is_correct'].mean().reset_index()

    rows = []
    for topic in sorted(overall.keys()):
        acc = overall[topic]
        tt = pta[pta['topic'] == topic]['is_correct'].tolist()
        if len(tt) < 3: continue
        std = np.std(tt)
        last = [round(a, 2) for a in tt[-G6_MAINT_MIN_TESTS:]]
        is_cold = momentum_states.get(topic) == 'cold_streak'

        if acc >= G6_MAINT_ACC_FLOOR and std < G6_MAINT_STD_CEIL and not is_cold:
            status = 'maintenance'
            stab = 'Highly Stable' if std < 0.08 else 'Stable'
            msg = ''
        elif acc >= G6_NOT_SAFE_ACC and (std >= G6_MAINT_STD_CEIL or is_cold):
            status = 'not_yet_safe'
            stab = 'Volatile'
            if is_cold:
                msg = 'Recent decline detected. Keep practicing until stable.'
            else:
                lo, hi = min(last), max(last)
                msg = (f'Scores bouncing ({lo:.0%} to {hi:.0%}). '
                       f'One more test above {G6_MAINT_ACC_FLOOR+0.03:.0%} → maintenance.')
        else:
            continue

        rows.append({'topic': topic, 'accuracy': round(acc, 3), 'last_tests': last,
                     'std': round(std, 3), 'stability': stab, 'status': status, 'message': msg})

    result = pd.DataFrame(rows)
    if not result.empty:
        so = {'maintenance': 0, 'not_yet_safe': 1}
        result['_s'] = result['status'].map(so)
        result = result.sort_values(['_s', 'accuracy'], ascending=[True, False]).drop(columns='_s')
    return result.reset_index(drop=True)


def compute_f17_study_plan(sr, df_master, maintenance_topics, prereq_graph,
                            momentum_states, overconfident_topics):
    """F17: Study focus plan — ordered priorities with daily time allocation."""
    topic_acc = compute_topic_accuracy(sr, df_master)
    # Exclude maintenance AND any topic ≥65% accuracy
    study = {t: a for t, a in topic_acc.items()
             if t not in maintenance_topics and a < G6_MAINT_ACC_FLOOR}
    if not study: return pd.DataFrame()

    weak_set = {t for t, a in topic_acc.items() if a < G3_F9_WEAK_TOPIC}
    prereq_depth = {}
    for t in study:
        dc = sum(1 for ds, ups in prereq_graph.items() if t in ups and ds in weak_set)
        prereq_depth[t] = dc

    rows = []
    for topic, acc in study.items():
        sev = 1 - acc
        pd_s = prereq_depth.get(topic, 0)
        mom = momentum_states.get(topic, 'unknown')
        stag = 1.0 if mom == 'stagnant' else (0.5 if mom == 'cold_streak' else 0.0)
        oc_type = overconfident_topics.get(topic, '')
        oc = 1.0 if oc_type in ('concept_overconfidence', 'mixed_overconfidence') else 0.0
        tip = 1.0 if G5_PHASE2_ACC <= acc < G5_PHASE1_ACC else 0.0

        pri = (sev * G6_SEVERITY_W + pd_s * G6_PREREQ_W + stag * G6_STAGNANT_W
               + oc * G6_OVERCONF_W + tip * G6_TIPPING_W)

        comps = []
        if sev > 0.5: comps.append(f'severity: {sev:.0%}')
        if pd_s > 0: comps.append(f'unlocks {pd_s} weak topic(s)')
        if stag > 0: comps.append('stagnant' if mom == 'stagnant' else 'regressing')
        if oc > 0: comps.append('overconfident blind spot')
        if tip > 0: comps.append('close to tipping into Quick Wins')

        cur = round(acc * 5)
        goal = f"Get {cur + 1} right instead of your usual {cur} (out of ~5)"

        rows.append({'topic': topic, 'accuracy': round(acc, 3), 'priority': round(pri, 2),
                     'momentum': mom, 'components': ', '.join(comps) or 'general weakness',
                     'micro_goal': goal})

    result = pd.DataFrame(rows)
    result = result.sort_values('priority', ascending=False).reset_index(drop=True)
    result['rank'] = range(1, len(result) + 1)

    total_budget = 90
    total_score = result['priority'].sum()
    if total_score > 0:
        result['daily_min'] = (result['priority'] / total_score * total_budget).round(0).astype(int).clip(lower=10)
    else:
        result['daily_min'] = total_budget // max(len(result), 1)
    return result


def compute_f18_next_action(sr, df_master, study_plan):
    """F18: Single next question from Priority 1 topic."""
    if study_plan.empty: return {}
    p1 = study_plan.iloc[0]
    tq = df_master[df_master['topic'] == p1['topic']].copy()
    seen = set(sr['question_id'])
    wrong_qids = set(sr[sr['is_correct'] == False]['question_id'])

    unseen = tq[~tq['question_id'].isin(seen)]
    prev_wrong = tq[tq['question_id'].isin(wrong_qids)]

    if not unseen.empty:
        sel = unseen.sort_values('text_length').iloc[len(unseen)//2]
        method = 'unseen'
    elif not prev_wrong.empty:
        sel = prev_wrong.sample(1, random_state=42).iloc[0]
        method = 'retry'
    else:
        sel = tq.sample(1, random_state=42).iloc[0]
        method = 'fallback'

    return {'topic': p1['topic'], 'accuracy': p1['accuracy'], 'reason': p1['components'],
            'micro_goal': p1['micro_goal'], 'rank': int(p1['rank']),
            'daily_min': int(p1['daily_min']),
            'question_id': sel['question_id'], 'question_type': sel['question_type'],
            'text_length': sel['text_length'], 'method': method}


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 9 — PIPELINE ORCHESTRATOR
# ═══════════════════════════════════════════════════════════════════════════════

@dataclass
class DiagnosticReport:
    """Container for all feature outputs."""
    student_id: str
    f1_node_weakness: pd.DataFrame = field(default_factory=pd.DataFrame)
    f2_concept_precision: pd.DataFrame = field(default_factory=pd.DataFrame)
    f3_error_patterns: pd.DataFrame = field(default_factory=pd.DataFrame)
    f4_failure_detail: pd.DataFrame = field(default_factory=pd.DataFrame)
    f4_failure_summary: pd.DataFrame = field(default_factory=pd.DataFrame)
    f5_overconfidence: pd.DataFrame = field(default_factory=pd.DataFrame)
    f6_underconfidence: pd.DataFrame = field(default_factory=pd.DataFrame)
    f7_demand_profile: pd.DataFrame = field(default_factory=pd.DataFrame)
    f7_drop_off: str = ''
    f8_classification: pd.DataFrame = field(default_factory=pd.DataFrame)
    f9_prerequisites: pd.DataFrame = field(default_factory=pd.DataFrame)
    f9_readiness: dict = field(default_factory=dict)
    f10_foundational: pd.DataFrame = field(default_factory=pd.DataFrame)
    f10_global_acc: float = 0.0
    f10_display: str = ''
    f11_momentum: dict = field(default_factory=dict)
    f14_exam_plan: dict = field(default_factory=dict)
    f15_time_roi: dict = field(default_factory=dict)
    f16_maintenance: pd.DataFrame = field(default_factory=pd.DataFrame)
    f17_study_plan: pd.DataFrame = field(default_factory=pd.DataFrame)
    f18_next_action: dict = field(default_factory=dict)


def run_diagnostic_engine(
    student_id: str,
    student_responses: pd.DataFrame,
    test_composition: pd.DataFrame,
    df_master: pd.DataFrame,
    df_question_nodes: pd.DataFrame,
    df_question_errors: pd.DataFrame,
    df_question_concepts: pd.DataFrame,
    prerequisite_graph: dict,
    foundational_skills: list,
) -> DiagnosticReport:
    """
    Run the full diagnostic pipeline for one student.
    Returns a DiagnosticReport with all feature outputs.
    """
    report = DiagnosticReport(student_id=student_id)

    # Filter to this student
    sr = student_responses[student_responses['student_id'] == student_id].copy()
    tc = test_composition[test_composition.get('student_id', student_id) == student_id] \
        if 'student_id' in test_composition.columns \
        else test_composition.copy()

    print(f"\n{'='*70}")
    print(f"DIAGNOSTIC ENGINE — Student: {student_id}")
    print(f"{'='*70}")
    print(f"  Responses: {len(sr)} | Tests: {sr['test_timestamp'].nunique()} | "
          f"Accuracy: {sr['is_correct'].mean():.0%}")

    # ── Stage 0: Pre-filter ───────────────────────────────────────────────────
    print("\n  Stage 0: Building eligible sets...")
    eligible_nodes, eligible_concepts, eligible_errors, node_lookup, error_lookup = \
        build_eligible_sets(df_question_nodes, df_question_errors, df_question_concepts)
    print(f"    Eligible: {len(eligible_nodes)} nodes, {len(eligible_concepts)} concepts, "
          f"{len(eligible_errors)} errors")

    # ── Stage 1: Independent features ─────────────────────────────────────────
    print("  Stage 1: F1, F2, F3, F4, F7, F8...")
    report.f1_node_weakness = compute_f1_node_weakness(sr, df_question_nodes, eligible_nodes, node_lookup)
    weak_nodes = report.f1_node_weakness[report.f1_node_weakness['status'] == 'weak']['node_name'].tolist()

    report.f2_concept_precision = compute_f2_concept_precision(
        sr, df_question_concepts, eligible_concepts, weak_nodes, df_question_nodes)

    report.f3_error_patterns = compute_f3_error_patterns(
        sr, df_question_errors, eligible_errors, error_lookup, df_master)

    report.f4_failure_detail, report.f4_failure_summary = compute_f4_failure_modes(
        sr, df_question_errors, df_master)

    report.f7_demand_profile, report.f7_drop_off = compute_f7_demand_profile(sr, df_question_nodes)

    report.f8_classification = compute_f8_classification(sr, df_question_nodes, df_master, node_lookup)

    # ── Stage 2: Confidence features ──────────────────────────────────────────
    print("  Stage 2: F5, F6...")
    report.f5_overconfidence = compute_f5_overconfidence(sr, df_question_nodes, df_question_errors, node_lookup)
    report.f6_underconfidence = compute_f6_underconfidence(sr, df_question_nodes, df_master, node_lookup)

    # ── Stage 3: Root cause + momentum ────────────────────────────────────────
    print("  Stage 3: F9, F10, F11...")
    report.f9_prerequisites, report.f9_readiness = compute_f9_prerequisites(
        sr, df_master, df_question_nodes, prerequisite_graph)

    report.f10_foundational, report.f10_global_acc, report.f10_display = compute_f10_foundational(
        sr, df_master, df_question_nodes, foundational_skills)

    report.f11_momentum = compute_f11_momentum(sr, df_master)

    # ── Stage 4: Strategy + maintenance ───────────────────────────────────────
    print("  Stage 4: F14, F16...")
    momentum_states = report.f11_momentum.get('momentum_states', {})

    report.f14_exam_plan = compute_f14_exam_plan(sr, df_master, df_question_nodes)
    report.f16_maintenance = compute_f16_maintenance(sr, df_master, momentum_states)

    # ── Stage 5: Study plan ───────────────────────────────────────────────────
    print("  Stage 5: F17...")
    maintenance_topics = set(
        report.f16_maintenance[report.f16_maintenance['status'] == 'maintenance']['topic']
    ) if not report.f16_maintenance.empty else set()

    # Extract overconfident topics from F5
    oc_flagged = report.f5_overconfidence[report.f5_overconfidence['flagged'] == True]
    overconfident_topics = {}
    if not oc_flagged.empty:
        # Map node-level overconfidence to topic level
        node_topic = df_question_nodes.merge(df_master[['question_id', 'topic']], on='question_id')
        for _, row in oc_flagged.iterrows():
            topics = node_topic[node_topic['node_name'] == row['node_name']]['topic'].unique()
            for t in topics:
                overconfident_topics[t] = row['diagnosis']

    report.f17_study_plan = compute_f17_study_plan(
        sr, df_master, maintenance_topics, prerequisite_graph,
        momentum_states, overconfident_topics)

    # ── Stage 6: Next action + time ROI ───────────────────────────────────────
    print("  Stage 6: F15, F18...")
    report.f15_time_roi = compute_f15_time_roi(sr, df_master, df_question_nodes, tc)
    report.f18_next_action = compute_f18_next_action(sr, df_master, report.f17_study_plan)

    print(f"\n  Pipeline complete. All features computed.")
    return report


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 10 — SUMMARY PRINTER
# ═══════════════════════════════════════════════════════════════════════════════

def print_report_summary(report: DiagnosticReport):
    """Print a compact summary of all feature outputs."""

    print(f"\n{'='*70}")
    print(f"DIAGNOSTIC REPORT SUMMARY — {report.student_id}")
    print(f"{'='*70}")

    # Group 1
    weak = report.f1_node_weakness[report.f1_node_weakness['status'] == 'weak']
    print(f"\n  GROUP 1 — Locating Weakness:")
    print(f"    F1  Weak nodes: {len(weak)}")
    for _, r in weak.head(3).iterrows():
        print(f"        → {r['node_name']} ({r['accuracy']:.0%})")
    f2_flagged = report.f2_concept_precision['flagged'].sum() if not report.f2_concept_precision.empty else 0
    print(f"    F2  Concepts flagged: {f2_flagged}")
    print(f"    F3  Recurring errors: {len(report.f3_error_patterns)}")

    # Group 2
    print(f"\n  GROUP 2 — Classifying Failure:")
    if not report.f4_failure_summary.empty:
        dom = report.f4_failure_summary.iloc[0]
        print(f"    F4  Dominant mode: {dom['failure_mode']} ({dom['pct']:.0f}%)")
    oc = report.f5_overconfidence[report.f5_overconfidence['flagged'] == True]
    print(f"    F5  Overconfident: {len(oc)} node(s)")
    uc = report.f6_underconfidence[report.f6_underconfidence['flagged'] == True]
    print(f"    F6  Underconfident: {len(uc)} node(s)")
    print(f"    F7  Demand drop-off: {report.f7_drop_off or 'None'}")

    # Group 3
    print(f"\n  GROUP 3 — Root Cause:")
    if not report.f8_classification.empty:
        c = report.f8_classification['classification'].value_counts().to_dict()
        print(f"    F8  Slips: {c.get('unexpected_slip',0)}, Stretch: {c.get('stretch_question',0)}, "
              f"Focus: {c.get('focus_area',0)}")
    if not report.f9_prerequisites.empty:
        conf = len(report.f9_prerequisites[report.f9_prerequisites['trace_confidence'] == 'confirmed'])
        print(f"    F9  Confirmed prereq traces: {conf}")
    print(f"    F10 Foundational hypotheses: {len(report.f10_foundational)}")

    # Group 4
    print(f"\n  GROUP 4 — Performance Momentum:")
    if report.f11_momentum:
        t = report.f11_momentum['trend']
        print(f"    F11 Trajectory: {t['slope']:+.1f} marks/test ({t['label']})")
        d = report.f11_momentum['drivers']
        for m in ['hot_streak', 'cold_streak', 'stagnant']:
            topics = d[d['momentum'] == m]['topic'].tolist()
            if topics: print(f"        {m}: {', '.join(topics)}")

    # Group 5
    print(f"\n  GROUP 5 — Exam Strategy:")
    if report.f14_exam_plan:
        proj = report.f14_exam_plan['projected']
        print(f"    F14 Projected: ~{proj['total']} marks "
              f"(P1: {proj['phase_1']}, P2: {proj['phase_2']}, P3: {proj['phase_3']})")
    if report.f15_time_roi:
        bl = report.f15_time_roi['bottom_line']
        recover = bl['with_tragic'] - bl['score']
        print(f"    F15 Today: {bl['score']} marks, recoverable: +{recover:.0f}")

    # Group 6
    print(f"\n  GROUP 6 — Action Plan:")
    maint = report.f16_maintenance
    mc = len(maint[maint['status'] == 'maintenance']) if not maint.empty else 0
    print(f"    F16 Maintenance topics: {mc}")
    if not report.f17_study_plan.empty:
        p1 = report.f17_study_plan.iloc[0]
        print(f"    F17 Priority 1: {p1['topic']} ({p1['accuracy']:.0%}, {int(p1['daily_min'])} min/day)")
    if report.f18_next_action:
        a = report.f18_next_action
        print(f"    F18 Next action: {a['question_id']} from {a['topic']}")
        print(f"        🎯 {a['micro_goal']}")

    print(f"\n{'='*70}\n")


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 11 — RUNNER (placeholder — replace with real data loading)
# ═══════════════════════════════════════════════════════════════════════════════

if __name__ == '__main__':
    print("=" * 70)
    print("JEE DIAGNOSTIC ENGINE — Unified Pipeline")
    print("=" * 70)
    print()
    print("To run this engine, load your data and call:")
    print()
    print("  report = run_diagnostic_engine(")
    print("      student_id       = 'S001',")
    print("      student_responses = df_student_responses,   # ← SYNTHETIC (to generate)")
    print("      test_composition  = df_test_composition,    # ← SYNTHETIC (to generate)")
    print("      df_master         = df_master,              # ← YOU HAVE THIS")
    print("      df_question_nodes = df_question_nodes,      # ← YOU HAVE THIS")
    print("      df_question_errors = df_question_errors,    # ← YOU HAVE THIS")
    print("      df_question_concepts = df_question_concepts,# ← YOU HAVE THIS")
    print("      prerequisite_graph = prerequisite_graph,     # ← FROM layer_1_PN.txt")
    print("      foundational_skills = foundational_skills,   # ← FROM prerequisite_final.json + labeling")
    print("  )")
    print()
    print("  print_report_summary(report)")
    print()
    print("INPUT FILES NEEDED:")
    print("─" * 50)
    print("  ALREADY HAVE:")
    print("    1. df_master            — 4481 questions (question_id, topic, question_type, text_length)")
    print("    2. df_question_nodes    — 5150 rows (question_id, node_name, cognitive_operation, cognitive_demand)")
    print("    3. df_question_errors   — 5522 rows (question_id, error_name, error_type, description)")
    print("    4. df_question_concepts — 11784 rows (question_id, topic, concept_name)")
    print("    5. layer_1_PN.txt       — 75 topic prerequisite edges → load as dict")
    print("    6. prerequisite_final.json — 42 foundational skills")
    print()
    print("  NEED TO ADD TO FOUNDATIONAL SKILLS (one-time labeling):")
    print("    7. cognitive_demand_signature — per skill, list of demand levels it maps to")
    print("       Example: F007 → ['Build Then Solve', 'Build and Work Backwards']")
    print("    8. diagnostic_question — per skill, one test question to verify hypothesis")
    print("       Example: F007 → 'A function f(x) is defined piecewise on [0,3]...")
    print()
    print("  NEED TO GENERATE (synthetic student data):")
    print("    9. student_responses — student_id, question_id, is_correct, time_taken_seconds,")
    print("                           confidence_rating, test_timestamp")
    print("   10. test_composition  — student_id, test_timestamp, question_id")
    print("                           (all questions in each test, including unattempted)")
    print()
```

```python

"""
NARRATIVE GENERATOR — Student-Facing Report Layer
═══════════════════════════════════════════════════════════════════════════

Transforms DiagnosticReport (raw data) into natural-language reports
that speak to the student like a mentor, not a calculator.

Four sections per report:
  1. WHERE YOU STAND       — F11 + F15 overall context
  2. WHAT'S HAPPENING      — F4 + F5 + F6 + F7 + F8 diagnostic
  3. WHERE THE LEVERAGE IS — F9 + F10 + F11 cold streaks + F17 reasoning
  4. WHAT TO DO            — F16 + F17 + F18 action plan

All 18 features have reframe rules applied.

Usage:
  narrative = generate_student_report(report, student_id)
  print(narrative)
"""

import pandas as pd
import numpy as np


# ═══════════════════════════════════════════════════════════════════════════════
# TRANSLATION TABLES (clinical labels → student language)
# ═══════════════════════════════════════════════════════════════════════════════

FAILURE_MODE_REFRAME = {
    'Concept Gap':     ('Foundation moments',  "you needed a stronger grip on the underlying idea"),
    'Setup Gap':       ('Starting line moments', "you knew the concept, but couldn't see the way in"),
    'Execution Slip':  ('Tiny moments',         "you had it — small mistakes cost you"),
    'Rushed':          ('Clock moments',        "time pressure got to you"),
}

DEMAND_LEVEL_REFRAME = {
    'Direct Application':       'plug-and-play questions',
    'Formula with Judgment':    'pick-the-right-formula questions',
    'Build Then Solve':         'construct-from-scratch questions',
    'Build and Work Backwards': 'reverse-engineer questions',
    'Reverse Engineer':         'puzzle questions',
}

F8_REFRAME = {
    'unexpected_slip':  ('Stuff you actually know', 'review what happened'),
    'stretch_question': ('Hard for everyone',       'not a slip, this is your ceiling'),
    'focus_area':       ('Known territory',         'we\'re working on this'),
    'developing':       ('In progress',             'could go either way'),
    'no_baseline':      ('First encounter',         'not enough history yet'),
}

OC_REFRAME = {
    'concept_overconfidence':
        "you trust your gut here, but the gut is wrong — restudy this",
    'execution_overconfidence':
        "you understand it, you just rush — slow down on these",
    'mixed_overconfidence':
        "a mix of concept gaps and careless slips — review both",
}

MOMENTUM_REFRAME = {
    'hot_streak':  'clicking',
    'cold_streak': 'slipping (rusty, not broken)',
    'stagnant':    'stuck',
    'stable':      'holding steady',
}


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 1 — WHERE YOU STAND
# ═══════════════════════════════════════════════════════════════════════════════

def _section_1_where_you_stand(report):
    """Opening section: trajectory + current score + recoverable marks."""
    lines = []
    lines.append("━" * 60)
    lines.append("  WHERE YOU STAND")
    lines.append("━" * 60)

    # Trajectory framing
    trend = report.f11_momentum.get('trend', {}) if report.f11_momentum else {}
    slope = trend.get('slope', 0)
    label = trend.get('label', 'Flat')
    current_avg = trend.get('current_avg', 0)

    if label == 'Steady Climb':
        traj_msg = (f"You're climbing — averaging {current_avg:.0f} marks across your last 5 tests, "
                    f"up {slope:+.1f} marks per test. Whatever you're doing, keep doing it.")
    elif label == 'Slight Upward':
        traj_msg = (f"You're moving in the right direction — {current_avg:.0f} marks average, "
                    f"trending up {slope:+.1f} per test. Not dramatic, but real.")
    elif label == 'Flat':
        traj_msg = (f"You're holding steady at {current_avg:.0f} marks. "
                    f"Not falling behind, but not accelerating either. "
                    f"The right adjustments could change this fast.")
    elif label == 'Slight Decline':
        traj_msg = (f"Your last 5 tests average {current_avg:.0f} marks, "
                    f"slightly down ({slope:+.1f} per test). "
                    f"Nothing alarming — but worth looking at what changed.")
    else:
        traj_msg = (f"Your scores have dropped to {current_avg:.0f} marks recently "
                    f"({slope:+.1f} per test). Let's figure out why and fix it.")

    lines.append("")
    lines.append(f"  {traj_msg}")

    # F15 recoverable marks
    if report.f15_time_roi and 'bottom_line' in report.f15_time_roi:
        bl = report.f15_time_roi['bottom_line']
        recoverable = bl.get('with_tragic', bl.get('score', 0)) - bl.get('score', 0)
        score = bl.get('score', 0)
        if recoverable > 0:
            lines.append("")
            lines.append(f"  Here's the unlock: you scored {score} on today's test, "
                         f"but {recoverable:.0f} of those missing marks are recoverable "
                         f"from strategy alone — no studying required.")
            lines.append(f"  We'll get to that.")

    return "\n".join(lines)


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 2 — WHAT'S ACTUALLY HAPPENING
# ═══════════════════════════════════════════════════════════════════════════════

def _section_2_whats_happening(report):
    """Diagnostic section: failure modes + blind spots + hidden strengths + demand."""
    lines = []
    lines.append("")
    lines.append("━" * 60)
    lines.append("  WHAT'S ACTUALLY HAPPENING")
    lines.append("━" * 60)
    lines.append("")

    # F4 dominant failure mode
    if not report.f4_failure_summary.empty:
        dom = report.f4_failure_summary.iloc[0]
        mode = dom['failure_mode']
        pct = dom['pct']
        reframe_label, reframe_desc = FAILURE_MODE_REFRAME.get(mode, (mode, ''))

        lines.append(f"  {pct:.0f}% of your wrong answers are \"{reframe_label}\" — ")
        lines.append(f"  {reframe_desc}. That's the dominant pattern in your mistakes.")

        # Add nuance based on mode
        if mode == 'Concept Gap':
            lines.append(f"  These need actual studying — not more practice, more understanding.")
        elif mode == 'Setup Gap':
            lines.append(f"  You're not weak on theory. You're weak on translating problems")
            lines.append(f"  into a solvable form. That's a separate skill, and it's fixable.")
        elif mode == 'Execution Slip':
            lines.append(f"  Good news: you know this material. Slowing down by 20% would")
            lines.append(f"  recover most of these marks.")
        elif mode == 'Rushed':
            lines.append(f"  This is pacing, not knowledge. We'll address it in your exam plan.")
        lines.append("")

    # F5 overconfidence (blind spots)
    oc = report.f5_overconfidence[report.f5_overconfidence['flagged'] == True] if not report.f5_overconfidence.empty else pd.DataFrame()
    if not oc.empty:
        n = len(oc)
        lines.append(f"  You have {n} blind spot{'s' if n > 1 else ''} — question types where")
        lines.append(f"  you feel solid but the data disagrees:")
        for _, row in oc.head(3).iterrows():
            diag = row['diagnosis']
            msg = OC_REFRAME.get(diag, '')
            lines.append(f"    → \"{row['node_name']}\" (you rate yourself {row['avg_conf']:.1f}/3, "
                         f"actual: {row['accuracy']:.0%})")
            if msg:
                lines.append(f"      {msg}")
        lines.append("")

    # F6 underconfidence (hidden strengths)
    uc = report.f6_underconfidence[report.f6_underconfidence['flagged'] == True] if not report.f6_underconfidence.empty else pd.DataFrame()
    if not uc.empty:
        n = len(uc)
        lines.append(f"  Flipside: {n} hidden strength{'s' if n > 1 else ''} — you're better here")
        lines.append(f"  than you think. Attempt these with confidence next time:")
        for _, row in uc.head(3).iterrows():
            diag = row.get('diagnosis', '')
            if diag == 'recently_corrected':
                continue   # already self-corrected
            lines.append(f"    → \"{row['node_name']}\" ({row['accuracy']:.0%} accuracy, "
                         f"but you rate yourself only {row['avg_conf']:.1f}/3)")
        lines.append("")

    # F7 cognitive demand drop-off
    if report.f7_drop_off:
        student_friendly = DEMAND_LEVEL_REFRAME.get(report.f7_drop_off, report.f7_drop_off)
        lines.append(f"  Where your accuracy drops off: {student_friendly}.")
        lines.append(f"  You handle simpler levels well, but once questions ask you to construct")
        lines.append(f"  your own approach, the wheels come off. That's the next level to unlock.")
        lines.append("")

    # F8 latest test breakdown
    if not report.f8_classification.empty:
        c = report.f8_classification['classification'].value_counts().to_dict()
        slip = c.get('unexpected_slip', 0)
        stretch = c.get('stretch_question', 0)
        focus = c.get('focus_area', 0)

        if slip > 0 or stretch > 0:
            lines.append(f"  On your latest test specifically:")
            if slip > 0:
                lines.append(f"    {slip} wrong answer{'s' if slip > 1 else ''} on stuff you usually get right — review these.")
            if stretch > 0:
                lines.append(f"    {stretch} wrong on genuinely hard questions — don't lose confidence here.")
            if focus > 0:
                lines.append(f"    {focus} on known weak areas — this is where the work continues.")
            lines.append("")

    return "\n".join(lines)


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 3 — WHERE THE LEVERAGE IS
# ═══════════════════════════════════════════════════════════════════════════════

def _section_3_where_leverage(report):
    """Causal section: prereq traces + foundational skill + momentum + Priority 1 reasoning."""
    lines = []
    lines.append("")
    lines.append("━" * 60)
    lines.append("  WHERE THE LEVERAGE IS")
    lines.append("━" * 60)
    lines.append("")

    has_content = False

    # F11 cold streaks (rust, not weakness)
    if report.f11_momentum and 'drivers' in report.f11_momentum:
        drivers = report.f11_momentum['drivers']
        cold = drivers[drivers['momentum'] == 'cold_streak'] if not drivers.empty else pd.DataFrame()
        hot = drivers[drivers['momentum'] == 'hot_streak'] if not drivers.empty else pd.DataFrame()

        if not hot.empty:
            top_hot = hot.head(3)['topic'].tolist()
            lines.append(f"  Three things are clicking right now: {', '.join(top_hot)}.")
            lines.append(f"  Whatever you're doing in these areas — keep doing it.")
            lines.append("")
            has_content = True

        if not cold.empty:
            top_cold = cold.head(2)
            lines.append(f"  Watch out: you're slipping on:")
            for _, row in top_cold.iterrows():
                bl_acc = row.get('bl_acc', 0)
                cu_acc = row.get('cu_acc', 0)
                lines.append(f"    → {row['topic']} ({bl_acc:.0%} → {cu_acc:.0%})")
            lines.append(f"  This is rust, not weakness. You used to nail these.")
            lines.append(f"  30 minutes of focused practice brings them back online.")
            lines.append("")
            has_content = True

    # F9 confirmed prerequisite traces
    if not report.f9_prerequisites.empty:
        confirmed = report.f9_prerequisites[report.f9_prerequisites['trace_confidence'] == 'confirmed']
        if not confirmed.empty:
            # Get unique weak_topic + best prereq pairs
            seen_topics = set()
            traces_shown = 0
            for _, row in confirmed.iterrows():
                if row['weak_topic'] in seen_topics or traces_shown >= 2:
                    continue
                if row['prereq_status'] != 'also_weak':
                    continue
                seen_topics.add(row['weak_topic'])
                traces_shown += 1
                lines.append(f"  There's a reason you're struggling with {row['weak_topic']}.")
                lines.append(f"  Your {row['prereq_topic']} foundation isn't strong enough yet,")
                lines.append(f"  and {row['weak_topic']} is built on it. Fix {row['prereq_topic']}")
                lines.append(f"  first, and {row['weak_topic']} will start unlocking automatically.")
                lines.append("")
                has_content = True

    # F10 foundational skill hypothesis
    if not report.f10_foundational.empty:
        top_skill = report.f10_foundational.iloc[0]
        conf = top_skill.get('confidence', '')
        weak_overlap = top_skill.get('weak_overlap', [])

        if conf in ('strong_hypothesis', 'moderate_hypothesis') and len(weak_overlap) >= 2:
            n_topics = len(weak_overlap)
            sample_topics = ', '.join(weak_overlap[:3])

            lines.append(f"  Here's something worth considering. We've noticed a pattern:")
            lines.append(f"  the same hidden skill appears to be holding back {n_topics} different")
            lines.append(f"  topics for you ({sample_topics}).")
            lines.append("")
            lines.append(f"  The skill: \"{top_skill['skill_name']}\".")
            lines.append(f"  {top_skill.get('description', '')}")
            lines.append("")
            lines.append(f"  This might not be {n_topics} separate problems. It might be one.")
            if top_skill.get('diagnostic_question'):
                lines.append(f"  Solve this one diagnostic question to find out:")
                lines.append(f"    \"{top_skill['diagnostic_question'][:120]}...\"")
                lines.append(f"  If you handle it easily → it's not the issue, keep working topic by topic.")
                lines.append(f"  If you struggle → we just found your highest-leverage fix.")
            lines.append("")
            has_content = True

    # F17 Priority 1 reasoning
    if not report.f17_study_plan.empty:
        p1 = report.f17_study_plan.iloc[0]
        components = p1.get('components', '')
        acc = p1.get('accuracy', 0)
        topic = p1.get('topic', '')

        # Determine the strategic reason
        if 'tipping point' in components.lower() or 'close to' in components.lower():
            reason = (f"You're at {acc:.0%} there — close enough that a small push moves it "
                      f"into your Quick Win zone. Once {topic} is a 70%+ topic for you, "
                      f"your overall test score jumps 3-4 marks per test, automatically. "
                      f"That's why it's Priority 1 — maximum leverage per minute.")
        elif 'unlocks' in components:
            reason = (f"It unlocks topics downstream of it. Fixing {topic} doesn't just "
                      f"help with {topic} — it helps with everything built on top of it.")
        elif 'stagnant' in components:
            reason = (f"You've been stuck here for multiple tests, which means your current "
                      f"approach isn't working. A change in strategy here will be visible "
                      f"on your next test.")
        elif 'overconfident' in components:
            reason = (f"You feel confident here but the marks aren't matching. Restudying "
                      f"this topic clears a blind spot that's been quietly costing you marks.")
        else:
            reason = f"It's where your effort produces the most marks per minute right now."

        lines.append(f"  Your highest-leverage topic this week is {topic}.")
        lines.append(f"  {reason}")
        lines.append("")
        has_content = True

    if not has_content:
        lines.append("  No critical leverage points identified — your current pattern is")
        lines.append("  consistent. Focus on steady practice across your weak areas.")
        lines.append("")

    return "\n".join(lines)


# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 4 — WHAT TO DO
# ═══════════════════════════════════════════════════════════════════════════════

def _section_4_what_to_do(report):
    """Action section: maintenance + study plan + next action."""
    lines = []
    lines.append("")
    lines.append("━" * 60)
    lines.append("  WHAT TO DO")
    lines.append("━" * 60)
    lines.append("")

    # F16 maintenance
    # if not report.f16_maintenance.empty:
    #     maint = report.f16_maintenance[report.f16_maintenance['status'] == 'maintenance']
    #     if not maint.empty:
    #         topics = maint['topic'].tolist()
    #         n = len(topics)
    #         lines.append(f"  Reclaim your study hours from these {n} stable topic{'s' if n > 1 else ''}:")
    #         for _, row in maint.iterrows():
    #             lines.append(f"    ✓ {row['topic']} ({row['overall_accuracy']:.0%}, {row['stability']})")
    #         lines.append("")
    #         lines.append(f"  Strategy: do NOT read the textbook for these this week. No videos,")
    #         lines.append(f"  no full revision sessions. Just solve 2 practice questions per week")
    #         lines.append(f"  to keep them fresh. Spend the rest of your time attacking weak zones.")
    #         lines.append("")


    if not report.f16_maintenance.empty:
        # Check if 'status' column exists, otherwise use the whole df if already filtered
        if 'status' in report.f16_maintenance.columns:
            maint = report.f16_maintenance[report.f16_maintenance['status'] == 'maintenance']
        else:
            maint = report.f16_maintenance

        if not maint.empty:
            topics = maint['topic'].tolist()
            n = len(topics)
            lines.append(f"  Reclaim your study hours from these {n} stable topic{'s' if n > 1 else ''}:")
            for _, row in maint.iterrows():
                # SAFELY GET ACCURACY AND STABILITY (This is where the fix is!)
                acc = row.get('overall_accuracy', row.get('accuracy', 0))
                stab = row.get('stability', 'stable')

                lines.append(f"    ✓ {row['topic']} ({acc:.0%}, {stab})")
            lines.append("")
            lines.append(f"  Strategy: do NOT read the textbook for these this week. No videos,")
            lines.append(f"  no full revision sessions. Just solve 2 practice questions per week")
            lines.append(f"  to keep them fresh. Spend the rest of your time attacking weak zones.")
            lines.append("")

    # F17 study plan
    if not report.f17_study_plan.empty:
        plan = report.f17_study_plan.head(3)   # top 3 only
        lines.append(f"  Your study time this week (~90 min/day total):")
        for _, row in plan.iterrows():
            rank = int(row['rank'])
            lines.append(f"    Priority {rank}: {row['topic']:35s} "
                         f"{int(row['daily_min'])} min/day  (currently {row['accuracy']:.0%})")
            lines.append(f"               🎯 {row['micro_goal']}")
        lines.append("")

    # F18 next action — the closer
    if report.f18_next_action:
        action = report.f18_next_action
        lines.append(f"  ┌{'─' * 56}┐")
        lines.append(f"  │  RIGHT NOW{' ' * 46}│")
        lines.append(f"  ├{'─' * 56}┤")
        lines.append(f"  │  Open this one question: {action['question_id']:<28s} │")
        lines.append(f"  │  From: {action['topic']:<48s} │")
        lines.append(f"  │{' ' * 56}│")
        lines.append(f"  │  Your goal: {action['micro_goal']:<43s}│")
        lines.append(f"  │{' ' * 56}│")
        lines.append(f"  │  Don't think about your overall score.{' ' * 17}│")
        lines.append(f"  │  Don't think about the other 25 topics.{' ' * 16}│")
        lines.append(f"  │  Just this one question.{' ' * 31}│")
        lines.append(f"  │{' ' * 56}│")
        lines.append(f"  │  If you solve it, you've done your job today.{' ' * 10}│")
        lines.append(f"  │  If you don't, that's also data. We'll adjust.{' ' * 9}│")
        lines.append(f"  └{'─' * 56}┘")
        lines.append("")

    return "\n".join(lines)


# ═══════════════════════════════════════════════════════════════════════════════
# MAIN GENERATOR
# ═══════════════════════════════════════════════════════════════════════════════

def generate_student_report(report, student_name=None):
    """
    Generate a complete student-facing report from a DiagnosticReport.

    Args:
        report: DiagnosticReport object from run_diagnostic_engine()
        student_name: Optional name for the header

    Returns:
        Multi-line string with 4 sections of narrative.
    """
    name = student_name or report.student_id
    header = []
    header.append("")
    header.append("═" * 60)
    header.append(f"  YOUR DIAGNOSTIC REPORT — {name.upper()}")
    header.append("═" * 60)

    sections = [
        "\n".join(header),
        _section_1_where_you_stand(report),
        _section_2_whats_happening(report),
        _section_3_where_leverage(report),
        _section_4_what_to_do(report),
    ]

    return "\n".join(sections)

# # ═══════════════════════════════════════════════════════════════════════════════
# # DEMO RUN
# # ═══════════════════════════════════════════════════════════════════════════════

# if __name__ == '__main__':
#     print("Narrative generator loaded. Usage:")
#     print()
#     print("  from narrative_generator import generate_student_report")
#     print("  ")
#     print("  for student_id, report in reports.items():")
#     print("      print(generate_student_report(report, student_id))")
#     print()
#     print("Or just paste this file into a new Colab cell after Cell 3.")

```
